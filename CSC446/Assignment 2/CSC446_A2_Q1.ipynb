```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
from scipy.stats import skew
n_small = 10000
n_big = 50000
```

```python
# a) Uniform (0,1)
uniform_data_small = np.random.uniform(0, 1, size=n_small)

#plotting
plt.hist(x=uniform_data_small, bins=50, density=True)
plt.show()

print("Mean: ", np.mean(uniform_data_small))
print("Variance: ", np.var(uniform_data_small))
print("Skewness: ",skew(uniform_data_small))
print("Kurtosis: ", kurtosis(uniform_data_small, fisher=False, bias=True))

#repeat the same process for 50000 sample size
uniform_data_big = np.random.uniform(0, 1, size=n_big)

#plotting
plt.hist(x=uniform_data_big, bins=100, density=True)
plt.show()

print("Mean: ", np.mean(uniform_data_big))
print("Variance: ", np.var(uniform_data_big))
print("Skewness: ",skew(uniform_data_big))
print("Kurtosis: ", kurtosis(uniform_data_big, fisher=False, bias=True))
```

```python
# b) exponential(mean=1)
expo_small = np.random.exponential(1, size=n_small)

#plotting
plt.hist(x=expo_small, bins=50, density=True)
plt.show()

print("Mean: ", np.mean(expo_small))
print("Variance: ", np.var(expo_small))
print("Skewness: ",skew(expo_small))
print("Kurtosis: ", kurtosis(expo_small, fisher=False, bias=True))

#repeat the same process for 50000 sample size
expo_big = np.random.exponential(1, size=n_big)

#plotting
plt.hist(x=expo_big, bins=50, density=True)
plt.show()

print("Mean: ", np.mean(expo_big))
print("Variance: ", np.var(expo_big))
print("Skewness: ",skew(expo_big))
print("Kurtosis: ", kurtosis(expo_big, fisher=False, bias=True))
```

```python
# c) Normal (mean=1,standard deviation=1).
normal_small = np.random.normal(1, 1, size=n_small)

#plotting
plt.hist(x=normal_small, bins=50, density=True)
plt.show()

print("Mean: ", np.mean(normal_small))
print("Variance: ", np.var(normal_small))
print("Skewness: ",skew(normal_small))
print("Kurtosis: ", kurtosis(normal_small, fisher=False, bias=True))

#repeat the same process for 50000 sample size
normal_big = np.random.normal(1, 1, size=n_big)

#plotting
plt.hist(x=normal_big, bins=50, density=True)
plt.show()

print("Mean: ", np.mean(normal_big))
print("Variance: ", np.var(normal_big))
print("Skewness: ",skew(normal_big))
print("Kurtosis: ", kurtosis(normal_big, fisher=False, bias=True))
```
