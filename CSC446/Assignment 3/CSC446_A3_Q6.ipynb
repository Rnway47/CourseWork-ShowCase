```python
import random
import math
import matplotlib.pyplot as plt

def expo_dist_random(mu, size):
    return [(-1/mu)*(math.log(random.random())) for x in range(size)]
```

```python
random.seed(1)
ran_sample = expo_dist_random(1, 1000) #mu=1, generate 1000 random numbers
#histogram with 11 bins from 0 to 5.5 with comparison to exponential cdf
expo = [(1 - math.exp(-random.random())) for x in range(1000)]
test_1 = plt.hist(expo, bins=[0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5], density=True, cumulative=True)
```
