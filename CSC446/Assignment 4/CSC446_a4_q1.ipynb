```python
import numpy
import random
import math
from matplotlib import pyplot as plt

#part a
numpy.random.seed(1)
n = 1000 #size
expo_sample = sorted(numpy.random.exponential(scale=1.0, size=n)) #generate and sort the exponential sample
estimate = [(x-0.5)/n for x in range(1, 1001)] # (i-0.5)/n
inverse_estimate = [-numpy.log(1-estimate[x]) for x in range(1000)] #F-1((i-0.5)/n)
plt.title("Q-Q plot")
plt.xlabel("sample of data from X") 
plt.ylabel("yj, estimate of the (j-0.5)/n quantile of X") 
plt.plot(inverse_estimate, expo_sample)
plt.show()

print("We can observe approximately a straight line with a slope of 1. As a result, F or the random exponential distribution is a member of an appropriate family of distributions with appropriate parameter values")
```

```python
#part b
#mean
sample_mean = sum(expo_sample) / n
print("Sample mean: ", sample_mean)
#variance
difference_from_mean = [(expo_sample[x]**2) for x in range(1000)]
sample_variance = (sum(difference_from_mean) - n*(sample_mean**2))/ (n-1)
print("Sample variance: ", sample_variance)

print("The sample mean roughly matches the sample variance when mu = 1 because sample mean is equal to 1/mu (scale factor) in theory and sample variance is equal to 1/(mu^2) in theory. It matches the exponential distribution characteristics.")
```

```python
#part c
k = 102  #number of class intervals, in between 1000^0.5 and 200, we will pick 102
s = 1 # # of parameters of the hypothesized distribution
d_o_f = k-s-1 # degree of freedom, used it to get X0
p = round(1 / k, 8) #round to 8 decimals to avoid invalid value for math.log()

Ei = n * p #expected frequency
#find the observed frequency
observed_list = [] #complete observed numbers in each ith class, Oi
for order in range(k):
    observed_list.append([-math.log(1-order*p) <= x < -math.log(1-((order+1)*p)) for x in expo_sample].count(True)) #count how many numbers are in each interval
diff_list = [((observed_list[x]-Ei)**2) / Ei for x in range(len(observed_list))]
x0 = sum(diff_list)
print("Value of x0:", x0)
critical_value = 124.3 #level of significance 0.05
print("Critical value for 100 degrees of freedom and 0.05 level of significance:", critical_value)
if x0 < critical_value:
    print("H0, the null hypothesis, is not rejected at the level of significance\n")
else:
    print("H0, the null hypothesis, is rejected at the 0.05 level of significance\n")
```
